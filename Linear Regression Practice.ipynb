### Extracting important features

```python
from sklearn.feature_selection import SelectKBest,f_regression
```

```python
featureSelector = SelectKBest(f_regression,k=4)
```

```python
featureSelector.fit(housing_df,price_df.Price)
```

```python
tf_feature_data = featureSelector.transform(housing_df)
```

```python
tf_feature_data.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX,testX, trainY,testY = train_test_split(tf_feature_data, price_df.Price)
```

```python
lr.fit(trainX,trainY)
```

```python
lr.score(testX,testY)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
y_pred = lr.predict(testX)
```

```python
mean_squared_error(y_pred=y_pred, y_true=testY)
```

```python
from sklearn.linear_model import RidgeCV,Ridge,LassoCV,ElasticNetCV
```

```python
rr = RidgeCV(alphas=[.8,9.0])
```

```python
rr.fit(trainX,trainY)
```

```python
y_pred = rr.predict(testX)
```

```python
mean_squared_error(y_pred=y_pred, y_true=testY)
```

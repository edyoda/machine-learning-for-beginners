```python
import pandas as pd
```

```python
service_request = pd.read_csv('service_request.csv', encoding='latin-1', parse_dates=['Closed Date','Opened Date'])
```

```python
service_request.info()
```

```python
service_request.head()
```

```python
service_request['ttr'] = service_request['Closed Date'] - service_request['Opened Date']
```

```python
service_request.head()
```

```python
service_request.drop(['Order #','Attachment Name','Type'], inplace=True, axis=1)
```

```python
service_request.drop(['Closed Date','Opened Date'], axis=1, inplace=True)
```

```python
service_request['Employee Login'].unique().shape
```

```python
service_request.drop(['Employee Login','Service Request #'], axis=1, inplace=True)
```

```python
service_request.drop(['Expected Serial #','Asset #'], axis=1, inplace=True)
```

```python
service_request.dropna(inplace=True)
```

```python
service_request.info()
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
varth = VarianceThreshold(threshold=0.1)
```

```python
feature_data = service_request.drop('ttr', axis=1)
```

```python
feature_tf = feature_data[['Employee Organization Name', 'Status', 'Description','Product Line', 'Activity Type', 'Activity SubType']]
```

```python
from sklearn_pandas import DataFrameMapper
from sklearn.preprocessing import LabelEncoder
```

```python
mapper = DataFrameMapper([
    ('Employee Organization Name', LabelEncoder()),
    ('Status', LabelEncoder()),
    ('Description', LabelEncoder()),
    ('Product Line', LabelEncoder()),
    ('Activity Type', LabelEncoder()),
    ('Activity SubType', LabelEncoder())
])
```

```python
df = mapper.fit_transform(feature_tf)
```

```python
df = varth.fit_transform(df)
```

```python
from sklearn.covariance import EllipticEnvelope
```

```python
out = EllipticEnvelope(contamination=.1)
```

```python
out.fit(df)
```

```python
outliers =  out.predict(df)
```

```python
df = pd.DataFrame(df)
```

```python
df['o'] = outliers
```

```python
df['o'] = df.o.map(lambda x: 0 if x == 1 else 1)
```

```python
s = service_request['ttr']
```

```python
target_data = s.map(lambda d: d.days +1 )
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf = RandomForestRegressor(n_estimators=5)
```

```python
rf.fit(df,target_data)
```

```python
rf.predict([[1,2,3,4]])
```

### Gettiung rows with non null Auth Capital

```python
data_non_na = data.loc[~data.AUTHORIZED_CAPITAL.isnull()]
```

```python
data_non_na['AUTHORIZED_CAPITAL_NUM'] = data_non_na.AUTHORIZED_CAPITAL.map(lambda x:int(x.replace(',','').replace('.00','')))
```

```python
data_non_na.info()
```

```python
data_non_na[['SUB_CATEGORY','AUTHORIZED_CAPITAL','AUTHORIZED_CAPITAL_NUM']].head()
```

#### Find top 5 highest auth caputal

```python
data_non_na.loc[data_non_na.AUTHORIZED_CAPITAL_NUM.sort_values(ascending=False).index[:5]]
```

```python
data_non_na_pc = data_non_na.loc[~data_non_na.PAIDUP_CAPITAL.isnull()]
```

```python
data_non_na_pc
```

```python
data_non_na_pc['PAIDUP_CAPITAL_NUM'] = data_non_na_pc.PAIDUP_CAPITAL.map(lambda x:float(x.replace(',','')))
```

```python
data_non_na_pc['DIFF'] = data_non_na_pc.AUTHORIZED_CAPITAL_NUM - data_non_na_pc.PAIDUP_CAPITAL_NUM
```

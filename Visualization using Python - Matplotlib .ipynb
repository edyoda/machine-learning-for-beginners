#### Hexagonal Bin Plot
* Useful alternative to scatter plots if data are too dense to plot each point individually

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
df['b'] = df['b'] + np.arange(1000)
```

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
```

* If the above data is represented in scatter plot as shown below, it's difficult to make inferences

```python
df.plot.scatter(x='a',y='b')
```

#### Subplotting

```python
df.plot(subplots=True, figsize=(6,6))
```

```python
fig, axes = plt.subplots(nrows=2, ncols=2, )
row = 2
col = 2
cols = df.columns.tolist()
col_ctr = 0

for i in range(row):
    for j in range(col):
        df[cols[col_ctr]].plot(ax=axes[i,j])
        axes[i,j].set_title(cols[col_ctr])
        col_ctr += 1


fig.show()
```

```python
fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_visible(False)
df = pd.DataFrame(np.random.rand(5, 3), columns=['a', 'b', 'c'])
df.plot(table=True, ax=ax)
```

```python
fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_visible(False)
df = pd.DataFrame(np.random.rand(5, 3), columns=['a', 'b', 'c'])
df.plot(table= np.round(df.T,2), ax=ax)
```

```python
fig, ax = plt.subplots(1, 1)
ax.get_xaxis().set_visible(False)
df = pd.DataFrame(np.random.rand(5, 3), columns=['a', 'b', 'c'])
df.plot(table= df, ax=ax)
```

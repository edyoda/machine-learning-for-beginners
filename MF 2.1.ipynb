```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'A':np.random.randint(1,10,20),
                   'B':np.random.randint(1,10,20)})
```

```python
df['C'] = df.A + df.B
```

```python
df
```

```python
df['D'] = pd.Series(0, index=df.index)
```

```python
def f(r):
    if r%2 == 0:
        return 'Even'
    else:
        return 'Odd'
    
df['E'] =df['C'].map(f)
```

```python
df.drop(['E'],axis=1,inplace=True)
```

```python
for col in df.columns:
    df[col+'_'] = df[col].map(f)
```

```python
df
```

```python
df = pd.DataFrame({'A':np.random.randint(1,10,20),
                   'B':np.random.randint(1,10,20),
                   'C':np.random.randint(1,10,20)})
```

```python
def f(r):
    if r['A'] > r['C']:
        return r['A']
    else:
        return r['C']

df['max'] = df.apply(f,axis=1)
```

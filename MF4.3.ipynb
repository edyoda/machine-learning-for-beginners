```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
import pandas as pd
```

```python
cols = ['age','workclass','fnlwgt','education','education-num','marital-status','occupation','relationship'
        ,'race','sex','capital-gain','capital-loss','hours-per-week','native-country','Salary']
income_data = pd.read_csv('Data/adult.data.txt',names=cols)
```

```python
income_data.info()
```

```python
cat_cols = ['workclass','education','marital-status','occupation','relationship','race','sex','native-country','Salary']
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in cat_cols:
    income_data[col] = le.fit_transform(income_data[col])
```

```python
income_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX,testX,trainY,testY = train_test_split(income_data.drop(['Salary'],axis=1), income_data['Salary'])
```

```python
knc = KNeighborsClassifier(n_neighbors=5)
```

```python
knc.fit(trainX,trainY)
```

```python
knc.score(testX,testY)
```

```python
knc.predict(testX[:4])
```

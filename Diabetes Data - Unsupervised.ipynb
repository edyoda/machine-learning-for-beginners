```python
### Finding all categorical columns
cat_cols = [col for col in diabetes_data.columns.tolist() if diabetes_data[col].dtype.char == 'O']
```

```python
from sklearn.preprocessing import LabelEncoder
for col in cat_cols:
    le = LabelEncoder()
    diabetes_data[col] = le.fit_transform(diabetes_data[col])
```

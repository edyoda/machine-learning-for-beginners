```python
import pandas as pd
```

```python
titanic_data = pd.read_csv('https://raw.githubusercontent.com/zekelabs/machine-learning-for-beginners/master/data/titanic-train.csv',index_col='PassengerId' )
```

```python
titanic_data.head()
```

### Get all the rows where age is missing

```python
titanic_data_age_na = titanic_data.loc[titanic_data.Age.isnull()] 
```

```python
titanic_data_age = titanic_data.loc[~titanic_data.Age.isnull()] 
```

```python
titanic_data_age.Age.mean()
```

```python
titanic_data_age.loc[titanic_data_age.Sex == 'male'].Age.mean()
```

```python
titanic_data_age.loc[titanic_data_age.Sex == 'female'].Age.mean()
```

### Fill all missing values by mean age in main table

```python
titanic_data.fillna({'Age':29.7})
```

```python
df1 = titanic_data_age_na.loc[titanic_data_age_na.Sex == "female"]
```

```python
df1['Age'] = pd.Series(27.9, df1.index)
```

```python
df1
```

```python
df2 = titanic_data_age_na.loc[titanic_data_age_na.Sex == "male"]
```

```python
df2['Age'] = pd.Series(30.7, df2.index)
```

```python
df3 = pd.concat([df1,df2])
```

```python
df_final = pd.concat([titanic_data_age,df3])
```

```python
df_final.info()
```

```python
df_final.dropna(subset=['Embarked'],inplace=True)
```
